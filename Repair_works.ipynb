```python
end_date = datetime.datetime.now()
start_date = end_date - datetime.timedelta(days=2*365)
```

```python
# data = pd.read_csv("Input/FNO_LIST.csv")
data = pd.read_csv(r"C:\Users\acer\Desktop\Py-Fin\Fin_data\Fin_data\Input\FNO_LIST.csv")
grouped = data.groupby('Sector')
resultList = []
for key, group in grouped:

    # Make it optimised in making pairs

    for b, g in combinations(group['SYMBOLS'].tolist(), 2):
        resultList.append([b, g, key])
        
resultList.sort()


total_list=resultList
total_list
```

```python
def Stockstore(stock1,stock2):
    
    """
    Downloads and stores stock closing data from yt finance
    """

    # print(stock1,stock2)
    data = yf.download(stock1+" "+stock2, start=start_date, end=end_date)
    # Take closing Value only
    data = data["Close"]
    data.fillna(method = 'ffill', inplace= True)
    data.fillna(method = 'bfill', inplace= True)
    data.to_csv(fr"C:\Users\acer\Desktop\Py-Fin\Fin_data\Fin_data\Data\Yt Stock Data\{stock1}_{stock2}_StoredData.csv")
```

```python
def cheg_list(total_list):
    
    '''
    Checks all pairs and returns list of pair with corr more than 0.84
    '''

    rlist=[]
    for all_list in total_list:
        s1=all_list[0]
        s2=all_list[1]
        
        # df = yf.download(s1+" "+s2, start=start_date, end=end_date)
        # df = df["Close"]
        df=pd.read_csv(fr"C:\Users\acer\Desktop\Py-Fin\Fin_data\Fin_data\Data\Yt Stock Data\{s1}_{s2}_StoredData.csv")
        cor = df[s1].corr(df[s2])
        if cor>0.84:
            rlist.append(all_list)
    return rlist
```
